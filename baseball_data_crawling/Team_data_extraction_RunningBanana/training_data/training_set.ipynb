## 위 코드는 값을 확인하기 위한 예제코드
## 아래코드는 머신러닝 학습에 필요한 csv 파일 제작을 위한 코드

```python
import numpy as np

for i in range(2020, 2021):
    pitcher_table = pd.read_csv('pitcher_processing' + str(i) + '.csv')
    hitter_table = pd.read_csv('hitter_processing' + str(i) + '.csv')
    statiz_table = pd.read_csv('statiz_game_data_with_entry_' + str(i) + '.csv', encoding='cp949')
    
    hitter_table['타수대비득점'].replace(np.inf, hitter_table['타수대비득점'].mean(axis=0), inplace=True)
    
    training_table = statiz_table.copy()
    # 기존의 column 기억하기 -> 나중에 drop
    ordinary_columns = list(training_table.columns)
    
    # 임시 열을 만드는 것, 원소값은 아래 반복문을 통해 수정한다
    # 투수쪽
    training_table['Team_1_pitcher_WHIP'] = 0
    training_table['Team_1_pitcher_piAn'] = 0
    training_table['Team_1_pitcher_piHom'] = 0
    training_table['Team_1_pitcher_sam'] = 0
    training_table['Team_1_pitcher_ball'] = 0
    training_table['Team_1_pitcher_ja'] = 0
    
    training_table['Team_2_pitcher_WHIP'] = 0
    training_table['Team_2_pitcher_piAn'] = 0
    training_table['Team_2_pitcher_piHom'] = 0
    training_table['Team_2_pitcher_sam'] = 0
    training_table['Team_2_pitcher_ball'] = 0
    training_table['Team_2_pitcher_ja'] = 0
    
    # 타자쪽
    training_table['Team_1_hitter_hit'] = 0
    training_table['Team_1_hitter_homrun'] = 0
    training_table['Team_1_hitter_sam'] = 0
    training_table['Team_1_hitter_ball'] = 0
    training_table['Team_1_hitter_deg'] = 0
    training_table['Team_1_hitter_ta'] = 0
    training_table['Team_1_hitter_chul'] = 0
    training_table['Team_1_hitter_do'] = 0
    
    training_table['Team_2_hitter_hit'] = 0
    training_table['Team_2_hitter_homrun'] = 0
    training_table['Team_2_hitter_sam'] = 0
    training_table['Team_2_hitter_ball'] = 0
    training_table['Team_2_hitter_deg'] = 0
    training_table['Team_2_hitter_ta'] = 0
    training_table['Team_2_hitter_chul'] = 0
    training_table['Team_2_hitter_do'] = 0
    
    training_table.drop('Date', axis=1, inplace=True) # 이름 열 삭제
    # 아래 행은 삭제 안할 행들이라 리스트에서 제거해줌. 이 리스트에 있는 행들은 drop할 것들.
    ordinary_columns.remove('Date')
    ordinary_columns.remove('Team_1_Name')
    ordinary_columns.remove('Team_2_Name')
    ordinary_columns.remove('Win_team')
    
    row_len = len(statiz_table.index)
    print(ordinary_columns)
    
    for idx in range(0, row_len):
        # 투수 데이터
        Team_1_pitcher_name = statiz_table['Team_1_pitcher_name'].iloc[idx]
        Team_2_pitcher_name = statiz_table['Team_2_pitcher_name'].iloc[idx]
        
        indexs1 = list(training_table[training_table['Team_1_pitcher_name'] == Team_1_pitcher_name].index)
        training_table.loc[indexs1, 'Team_1_pitcher_WHIP'] = pitcher_table[pitcher_table['이름']==Team_1_pitcher_name].iloc[0]['WHIP']
        training_table.loc[indexs1, 'Team_1_pitcher_piAn'] = pitcher_table[pitcher_table['이름']==Team_1_pitcher_name].iloc[0]['피안타율']
        training_table.loc[indexs1, 'Team_1_pitcher_piHom'] = pitcher_table[pitcher_table['이름']==Team_1_pitcher_name].iloc[0]['피홈런율']
        training_table.loc[indexs1, 'Team_1_pitcher_sam'] = pitcher_table[pitcher_table['이름']==Team_1_pitcher_name].iloc[0]['이닝당삼진']
        training_table.loc[indexs1, 'Team_1_pitcher_ball'] = pitcher_table[pitcher_table['이름']==Team_1_pitcher_name].iloc[0]['이닝당볼넷']
        training_table.loc[indexs1, 'Team_1_pitcher_ja'] = pitcher_table[pitcher_table['이름']==Team_1_pitcher_name].iloc[0]['평균자책점']
        
        indexs2 = list(training_table[training_table['Team_2_pitcher_name'] == Team_2_pitcher_name].index)
        training_table.loc[indexs2, 'Team_2_pitcher_WHIP'] = pitcher_table[pitcher_table['이름']==Team_2_pitcher_name].iloc[0]['WHIP']
        training_table.loc[indexs2, 'Team_2_pitcher_piAn'] = pitcher_table[pitcher_table['이름']==Team_2_pitcher_name].iloc[0]['피안타율']
        training_table.loc[indexs2, 'Team_2_pitcher_piHom'] = pitcher_table[pitcher_table['이름']==Team_2_pitcher_name].iloc[0]['피홈런율']
        training_table.loc[indexs2, 'Team_2_pitcher_sam'] = pitcher_table[pitcher_table['이름']==Team_2_pitcher_name].iloc[0]['이닝당삼진']
        training_table.loc[indexs2, 'Team_2_pitcher_ball'] = pitcher_table[pitcher_table['이름']==Team_2_pitcher_name].iloc[0]['이닝당볼넷']
        training_table.loc[indexs2, 'Team_2_pitcher_ja'] = pitcher_table[pitcher_table['이름']==Team_2_pitcher_name].iloc[0]['평균자책점']
        
        # 타자 데이터
        Team_1_Lineup = statiz_table['Team_1_Lineup'].iloc[idx]
        Team_2_Lineup = statiz_table['Team_2_Lineup'].iloc[idx]
        
        # 마지막 원소는 공백으로 삭제, 스페이스바로 구분된 타자들 리스트 형태로
        Team_1_Lineup = Team_1_Lineup.split(' ')[:-1]
        Team_2_Lineup = Team_2_Lineup.split(' ')[:-1]
        
        average_hit = 0.0
        average_homrun = 0.0
        average_sam = 0.0
        average_ball = 0.0
        average_deg = 0.0
        average_ta = 0.0
        average_chul = 0.0
        average_do = 0.0
        
        for name in Team_1_Lineup:
            #print(type(hitter_table[hitter_table['이름']==name].iloc[0].loc['타율']))
            print(name, i)
            average_hit += hitter_table[hitter_table['이름']==name].iloc[0].loc['타율']
            average_homrun += hitter_table[hitter_table['이름']==name].iloc[0].loc['타수대비홈런']
            average_sam += hitter_table[hitter_table['이름']==name].iloc[0].loc['타수대비삼진']
            average_ball += hitter_table[hitter_table['이름']==name].iloc[0].loc['타수대비사구']
            average_deg += hitter_table[hitter_table['이름']==name].iloc[0].loc['타수대비득점']
            average_ta += hitter_table[hitter_table['이름']==name].iloc[0].loc['타수대비타점']
            average_chul += hitter_table[hitter_table['이름']==name].iloc[0].loc['출루율']
            average_do += hitter_table[hitter_table['이름']==name].iloc[0].loc['도루성공률']
        
        average_hit /= len(Team_1_Lineup)
        average_homrun /= len(Team_1_Lineup)
        average_sam /= len(Team_1_Lineup)
        average_ball /= len(Team_1_Lineup)
        average_deg /= len(Team_1_Lineup)
        average_ta /= len(Team_1_Lineup)
        average_chul /= len(Team_1_Lineup)
        average_do /= len(Team_1_Lineup)
        
        training_table['Team_1_hitter_hit'].iloc[idx] = average_hit
        training_table['Team_1_hitter_homrun'].iloc[idx] = average_homrun
        training_table['Team_1_hitter_sam'].iloc[idx] = average_sam
        training_table['Team_1_hitter_ball'].iloc[idx] = average_ball
        training_table['Team_1_hitter_deg'].iloc[idx] = average_deg
        training_table['Team_1_hitter_ta'].iloc[idx] = average_ta
        training_table['Team_1_hitter_chul'].iloc[idx] = average_chul
        training_table['Team_1_hitter_do'].iloc[idx] = average_do
        
        average_hit = 0.0
        average_homrun = 0.0
        average_sam = 0.0
        average_ball = 0.0
        average_deg = 0.0
        average_ta = 0.0
        average_chul = 0.0
        average_do = 0.0
        
        for name in Team_2_Lineup:
            #print(type(hitter_table[hitter_table['이름']==name].iloc[0].loc['타율']))
            print(name, i)
            average_hit += hitter_table[hitter_table['이름']==name].iloc[0].loc['타율']
            average_homrun += hitter_table[hitter_table['이름']==name].iloc[0].loc['타수대비홈런']
            average_sam += hitter_table[hitter_table['이름']==name].iloc[0].loc['타수대비삼진']
            average_ball += hitter_table[hitter_table['이름']==name].iloc[0].loc['타수대비사구']
            average_deg += hitter_table[hitter_table['이름']==name].iloc[0].loc['타수대비득점']
            average_ta += hitter_table[hitter_table['이름']==name].iloc[0].loc['타수대비타점']
            average_chul += hitter_table[hitter_table['이름']==name].iloc[0].loc['출루율']
            average_do += hitter_table[hitter_table['이름']==name].iloc[0].loc['도루성공률']
        
        average_hit /= len(Team_2_Lineup)
        average_homrun /= len(Team_2_Lineup)
        average_sam /= len(Team_2_Lineup)
        average_ball /= len(Team_2_Lineup)
        average_deg /= len(Team_2_Lineup)
        average_ta /= len(Team_2_Lineup)
        average_chul /= len(Team_2_Lineup)
        average_do /= len(Team_2_Lineup)
        
        training_table['Team_2_hitter_hit'].iloc[idx] = average_hit
        training_table['Team_2_hitter_homrun'].iloc[idx] = average_homrun
        training_table['Team_2_hitter_sam'].iloc[idx] = average_sam
        training_table['Team_2_hitter_ball'].iloc[idx] = average_ball
        training_table['Team_2_hitter_deg'].iloc[idx] = average_deg
        training_table['Team_2_hitter_ta'].iloc[idx] = average_ta
        training_table['Team_2_hitter_chul'].iloc[idx] = average_chul
        training_table['Team_2_hitter_do'].iloc[idx] = average_do
        
        # training_table.drop(ordinary_columns, axis=1, inplace=True)
        training_table.to_csv('training_data_'+str(i)+'.csv', encoding='utf-8-sig')
```

```python
for i in range(2017, 2021):
    table = pd.read_csv('training_data_' + str(i) + '.csv')
    
    drop_columns = ['Team_1_Name', 'Team_2_Name', 'Team_1_pitcher_name', 'Team_2_pitcher_name', 'Team_1_Lineup', 'Team_2_Lineup', 'Team_1_Score', 'Team_2_Score', 'Unnamed: 0']
    table.drop(drop_columns, axis=1, inplace=True)
    print(table.columns)
    
    table.to_csv('training_data_drop' + str(i) + '.csv', encoding='utf-8-sig')
```
