```python
from selenium import webdriver
import csv

def crawl_data(year, month):
    print(str(year)+"/"+str(month))
    
    #setting CSV writer
    column_list = ["Date","Team_1_Name","Team_1_Score","Team_2_Name","Team_2_Score"]
    with open("statiz_game_data_"+str(year)+"_"+str(month)+".csv",'w',-1,newline='') as f:
        w = csv.writer(f)
        w.writerow(column_list)

        chromedriver_path = './chromedriver'
        url = "http://www.statiz.co.kr/schedule.php?opt="+str(month)+"&sy="+str(year)

        #Turn on Chrome
        driver = webdriver.Chrome(chromedriver_path)
        driver.get(url)


        table = driver.find_element_by_xpath("/html/body/div/div[1]/div/section[2]/div/div[2]/div/div/div/div[2]/table")
        trs = table.find_elements_by_tag_name("tr")
        for trData in trs:
            tds = trData.find_elements_by_tag_name("td")
            for tdData in tds:
                divs = tdData.find_elements_by_tag_name("div")
                if len(divs)!= 0:
                    print("경기일 : "+divs[0].text)
                    atags = divs[3].find_elements_by_tag_name("a")
                    for aData in atags:
                        spans = aData.find_elements_by_tag_name("span")
                        if len(spans) == 4:
                            row_list =[]
                            row_list.append(str(year)+"/"+str(month)+"/"+divs[0].text)
                            row_list.append(spans[0].text)
                            row_list.append(spans[1].text)
                            row_list.append(spans[3].text)
                            row_list.append(spans[2].text)
                            w.writerow(row_list)
                            #print("팀1: " + spans[0].text)
                            #print("팀1점수: " + spans[1].text)
                            #print("팀2점수: " + spans[2].text)
                            #print("팀2: " + spans[3].text)
```

```python
for m in range(3,11):
    crawl_data(2019,m)
    print("Job Finish!")
```
